```python
import numpy as np
import scipy
from scipy.stats import wasserstein_distance

# constraint: n â‰¥ 3
n = np.random.randint(3, 10)

A = np.random.randint(0, 10, size = (n, n))
B = np.random.randint(0, 10, size = (n, n))

# C_T = np.transpose(C)
# A = np.transpose(C)*C

# bures-wasserstein: [tr(A) + tr(B) - 2(A^1/2*B*A^1/2)^1/2]^1/2

trace_A = np.trace(A)
trace_B = np.trace(B)

# use sqrtm for root of matrix
# use real numbers to remove imaginary eigen components
A_root = scipy.linalg.sqrtm(A).real
trace_A_B = 2 * np.trace(A_root*B*A_root)
sum_trace = trace_A + trace_B + trace_A_B

# use real numbers to remove imaginary eigen components
d_BW = np.sqrt(sum_trace).real
d_BW_real = d_BW.real
print(d_BW_real, "d_BW_real")

# d2(A,B): [tr(A) = tr(B) - 2 \Sigma_i^n \sqrt(lambda_i)]^1/2
# where lambda_i are the eigenvalues of A*B

AB = A*B
lambda_ab = np.linalg.eig(AB)

# ignore first index of tuple'd matrix
lambda_mat = lambda_ab[1]

trace_root_lambda = np.trace(lambda_mat).real
d2_A_B = np.sqrt(trace_A + trace_B + trace_root_lambda)
print(d2_A_B, "d2_A_B")

# root_lambda_list = []
# for i in range(n):
#   root_lambda = np.sqrt(abs(lambda_ab[i]))
#   root_lambda_list.append(root_lambda)
# print(root_lambda_list, "root_lambda_list")

# simple verification using scipy wasserstein distance; loop indexes
d_BW_list = []
for i in range(n):
  d_BW = wasserstein_distance(A[i], B[i])
  d_BW_list.append(d_BW)
print(d_BW_list)
```
